## 01 – Dados as operações abaixo, em Python, qual o resultado da operação lógica: 
p and q and r and s and t?
<br>
<br>Imprima os objetos e o resultado usando uma estrutura condicional.

<br>p = 2.0 + 3.0 / 5.0 == 10
<br>q = (p + 10.2) - 5.0 * 2.0 <= 2.9
<br>r = 10 <= p / q + 33.8 < 100
<br>s = 5 != p / (q + r)
<br>t = 33 == s * 2

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

resultado = p and q and r and s and t

print("p:", p)
print("q:", q)
print("r:", r)
print("s:", s)
print("t:", t)

if resultado:
    print("Resultado final: True")
else:
    print("Resultado final: False")
```

## 04 – Escreva o programa em Python:

```python
aleatorios = np.random.normal(loc=10, scale=3, size=5)

print("Array gerado:", aleatorios)

valores_maiores_10 = aleatorios[aleatorios > 10]

quantidade = len(valores_maiores_10)

print("Quantidade de valores > 10:", quantidade)
print("Valores > 10:", valores_maiores_10)
```

```python
df_aula10.index = df_aula10.index.astype(str)
tamanho_genes = df_aula10.index.str.len()
print(tamanho_genes)

primeiro_gene = df_aula10.index[0]
tam_primeiro = len(primeiro_gene)
print(primeiro_gene, tam_primeiro)


if 2 <= tam_primeiro <= 4:
    print("Informações do gene:")
    print(df_aula10.loc[primeiro_gene])
elif tam_primeiro < 2:
    genes_menor_2 = df_aula10[tamanho_genes < 2]
    print("Genes com menos de 2 caracteres:")
    print(genes_menor_2)
else: tam_primeiro > 4
print("Genes com nome maior que 4 caracteres:")
print(df_aula10[tamanho_genes > 4])
```
