## 10 – Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe, três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas . Recupere o dicionário.

```python
df_csv = pd.read_csv("data.csv")


num_entradas = len(df_csv)
classes_colunas = list(df_csv.dtypes)
contagem_classes = df_csv.dtypes.value_counts().tolist()
colunas = [df_csv.iloc[:,0], df_csv.iloc[:,3], df_csv.iloc[:,8]]
medias = [col.mean() for col in colunas]
desvios = [col.std() for col in colunas]
terceiro_quartil = [col.quantile(0.75) for col in colunas]

dicio_df_csv = {
    "num_entradas": num_entradas,
    "classes_colunas": classes_colunas,
    "contagem_classes": contagem_classes,
    "medias": medias,
    "desvios_padrao": desvios,
    "terceiro_quartil": terceiro_quartil
}

print(dicio_df_csv)
```
