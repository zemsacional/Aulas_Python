## 03 – Escreva o programa em Python:


```python
lista_cinco_mil = list(range(5000))

pares = [x for x in lista_cinco_mil if x % 2 == 0]
impares = [x for x in lista_cinco_mil if x % 2 != 0]

print("Pares:", len(pares))
print("Ímpares:", len(impares))
```

```python
presenca_hu = df_aula10.columns.str.contains("hu", case=False)

for i, gene in enumerate(df_aula10.columns):
    print(f"Índice: {i}")
    print(f"Gene: {gene}")
    print(f"Tem 'hu'? {presenca_hu[i]}")
    print(f"Primeiros 2 caracteres: {gene[:2]}")
    print("-" * 30)

#ps: o último print eu pedi ao chat, pois estava muito difícil enxergar o resultado kkkk
```
