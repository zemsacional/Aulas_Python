## 09 – Instancie cinco variáveis da classe lista. Cada uma delas será referente à um dos valores contidos na lista instanciada no exercício 4 (Nomes dos genes). Em cada uma das listas, você deve inserir o número de caracteres presentes no nome do gene, os dois primeiros caracteres do nome, quantas vezes aparece o caractere ‘a’ e se no nome aparecem os caracteres ‘hu’. Para isso você deve utilizar a indexação de lista. Retorne cada um dos objetos:

```python
genes = ["TP53", "TNF", "IL6", "IGF2", "BRCA1"]

lista1 = [
    len(genes[0]),                 
    genes[0][:2],              
    genes[0].lower().count('a'), 
    'hu' in genes[0].lower()
]

lista2 = [
    len(genes[1]),
    genes[1][:2],
    genes[1].lower().count('a'),
    'hu' in genes[1].lower()
]

lista3 = [
    len(genes[2]),
    genes[2][:2],
    genes[2].lower().count('a'),
    'hu' in genes[2].lower()
]

lista4 = [
    len(genes[3]),
    genes[3][:2],
    genes[3].lower().count('a'),
    'hu' in genes[3].lower()
]

lista5 = [
    len(genes[4]),
    genes[4][:2],
    genes[4].lower().count('a'),
    'hu' in genes[4].lower()
]


print(lista1)
print(lista2)
print(lista3)
print(lista4)
print(lista5)
```

## 10 – Instancie uma variável da classe dicionário, os index dos valores devem ser os valores contidos na lista instanciada no exercício 4 (Nomes dos genes). Para isso você deve utilizar a indexação de lista. Os valores devem ser as listas instanciadas no exercício 9. Retorne o objeto.

```python
genes = ["TP53", "TNF", "IL6", "IGF2", "BRCA1"]

lista1 = [
    len(genes[0]),                 
    genes[0][:2],              
    genes[0].lower().count('a'), 
    'hu' in genes[0].lower()
]

lista2 = [
    len(genes[1]),
    genes[1][:2],
    genes[1].lower().count('a'),
    'hu' in genes[1].lower()
]

lista3 = [
    len(genes[2]),
    genes[2][:2],
    genes[2].lower().count('a'),
    'hu' in genes[2].lower()
]

lista4 = [
    len(genes[3]),
    genes[3][:2],
    genes[3].lower().count('a'),
    'hu' in genes[3].lower()
]

lista5 = [
    len(genes[4]),
    genes[4][:2],
    genes[4].lower().count('a'),
    'hu' in genes[4].lower()
]


dicionario = {
    genes[0]: lista1,
    genes[1]: lista2,
    genes[2]: lista3,
    genes[3]: lista4,
    genes[4]: lista5
}

print(dicionario)
```
