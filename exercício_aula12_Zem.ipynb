## 08 – Escreva o programa em Python:

```python
def calcular_media(valores_insumos):
    if len(valores_insumos) == 0:
        return "Erro: a lista está vazia."
    if not all(isinstance(x, (int, float)) for x in valores_insumos):
        return "Erro: todos os valores devem ser numéricos."
    media = sum(valores_insumos) / len(valores_insumos)
    return media

valores_insumos = [27, 487, 72, 168, 346]

resultado = calcular_media(valores_insumos)
print("Média:", resultado)
```

## 09 – Desenvolva um programa de computador que retorne a divisão de dois objetos simples. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def pypy_simples(a, b):
    print("Classe de 'a':", type(a))
    print("Classe de 'b':", type(b))
    
    print("Valor de 'a':", a)
    print("Valor de 'b':", b)
    
    if not isinstance(a, (int, float)) or not isinstance(b, (int, float)):
        return "Erro: ambos os objetos devem ser numéricos."
    if b == 0:
        return "Erro: divisão por zero não é permitida."
    
    resultado = a / b

    return resultado


# Exemplo:
res = pypy_simples(20, 4)
print("Resultado:", res)
```

## 10 – Desenvolva um programa de computador que retorne a divisão de dois objetos compostos. Seu programa deve ser capaz de reconhecer a classe destes objetos e o valor, antes de retornar ao usuário.

```python
def pypy_maluco(a, b):
    print("Classe de 'a':", type(a))
    print("Classe de 'b':", type(b))
    
    print("Valor de 'a':", a)
    print("Valor de 'b':", b)
    
    if not hasattr(a, "__iter__") or not hasattr(b, "__iter__"):
        return "Erro: ambos devem ser objetos compostos (iteráveis)."
    if len(a) != len(b):
        return "Erro: os objetos devem ter o mesmo tamanho."
    
    resultado = []
    
    for x, y in zip(a, b):
        if not isinstance(x, (int, float)) or not isinstance(y, (int, float)):
            return "Erro: todos os elementos devem ser numéricos."
        
        if y == 0:
            return "Erro: divisão por zero detectada."
        
        resultado.append(x / y)
    
    return resultado


# Exemplo:
a = [32, 58, 50]
b = [2, 4, 5]

res = pypy_maluco(a, b)
print("Resultado:", res)
```
