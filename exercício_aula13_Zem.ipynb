## 06 – Escreva o programa em Python:

```python
pd.read_csv("df_aula06.csv")

df_uso_excessivo = pd.read_csv("df_aula06.csv")

print(df_uso_excessivo)
```

```python
tamanhos = df_uso_excessivo.columns.astype(str).str.len()

freq = tamanhos.value_counts().sort_index()

print(freq)
```

```python
tabela_freq = freq.reset_index()
tabela_freq.columns = ["num_caracteres", "frequencia"]

print(tabela_freq)
```

```python
import matplotlib.pyplot as plt

tabela_freq.plot(
    x="num_caracteres",
    y="frequencia",
    kind="bar"
)

plt.xlabel("Número de caracteres")
plt.ylabel("Frequência")
plt.title("Frequência do tamanho dos genes")

plt.show()
```

## 07 – Reorganize a DataFrame, removendo as colunas contendo os dois primeiros caracteres, o número de caracter 'a', e a presença de 'hu', dos genes. Renomeie as colunas de acordo com as informações. Adicione uma coluna com a informação de 'Amostra 1'.

```python
colunas_remover = ["primeiros_2", "contagem_a", "tem_hu"]

df_uso_excessivo = df_uso_excessivo.drop(columns=colunas_remover, errors="ignore")

df_uso_excessivo.columns = [f"Gene_{i+1}" for i in range(len(df_uso_excessivo.columns))]

df_uso_excessivo["Amostra 1"] = 1

print(df_uso_excessivo.head())
```

## 08 – Usando a função randint(), da classe random do módulo numpy (np.random.randint(1,10,5)), instancie objeto da classe numpy array. Calcule as frequências desta nova amostra, considerando que à uma ordem entre os genes. Insira na DataFrame este objeto como "Amostra 2".

```python
#precisei colocar 4 ao invés de 5, graças ao tamanho do nosso df.
amostra_02 = np.random.randint(1, 10, 4)

print("Amostra 02:", amostra_02)

freq_amostra_02 = pd.Series(amostra_02).value_counts().sort_index()

print(freq_amostra_02)

df_uso_excessivo["Amostra 2"] = amostra_02

df_freq = freq_amostra_02.reset_index()
df_freq.columns = ["valor", "frequencia"]

print(df_freq)
```
